### –î–æ–º–∞—à–Ω–µ–µ –∑–∞–¥–∞–Ω–∏–µ "–ü—Ä–æ–∏–∑–≤–æ–¥–Ω–∞—è —Ñ—É–Ω–∫—Ü–∏–∏ –Ω–µ—Å–∫–æ–ª—å–∫–∏—Ö –∞—Ä–≥—É–º–µ–Ω—Ç–æ–≤".

```python
import numpy as np
from scipy.optimize import approx_fprime
```

–£—Ä–æ–≤–µ–Ω—å 0:
    
–ü–æ—Å—á–∏—Ç–∞–π—Ç–µ —á–∞—Å—Ç–Ω—ã–µ –ø—Ä–æ–∏–∑–≤–æ–¥–Ω—ã–µ —Ñ—É–Ω–∫—Ü–∏–π:

1) $f(x,y)=2x^2y^3 + 1/x + y^2x + 7$, –∞ –¥–ª—è —ç—Ç–æ–π —Ñ—É–Ω–∫—Ü–∏–∏ —Ç–∞–∫–∂–µ –≥—Ä–∞–¥–∏–µ–Ω—Ç –≤ —Ç–æ—á–∫–µ $(1,2)$

2) $f(x,y)=x^2y - sin(xy) + cos(x^2) + 6y$

```python
# –ü–æ—Å—á–∏—Ç–∞–π—Ç–µ —á–∞—Å—Ç–Ω—ã–µ –ø—Ä–æ–∏–∑–≤–æ–¥–Ω—ã–µ —Ñ—É–Ω–∫—Ü–∏–π: 1)  ùëì(ùë•,ùë¶)=2ùë•2ùë¶3+1/ùë•+ùë¶2ùë•+7 , –∞ –¥–ª—è —ç—Ç–æ–π —Ñ—É–Ω–∫—Ü–∏–∏ —Ç–∞–∫–∂–µ –≥—Ä–∞–¥–∏–µ–Ω—Ç –≤ —Ç–æ—á–∫–µ  (1,2)
def f1(v,c0,c1,c2,c3):
    return (c0 * v[0]**2*v[1]**3) + c1 * 1/v[0] + (c2 * v[1]**2*v[0]) + c3

# c0 = 2
# c1 = 1
# c2 = 1
# c3 = 7
```

```python
from sympy import *
x, y = symbols('x y')

print(f'–ß–∞—Å—Ç–Ω–∞—è –ø—Ä–æ–∏–∑–≤–æ–¥–Ω–∞—è –ø–æ —Ö: {diff((2*x**2*y**3) + 1/x + (y**2*x) + 7, x)}') #–ß–∞—Å—Ç–Ω–∞—è –ø—Ä–æ–∏–∑–≤–æ–¥–Ω–∞—è –ø–æ —Ö
print(f'–ß–∞—Å—Ç–Ω–∞—è –ø—Ä–æ–∏–∑–≤–æ–¥–Ω–∞—è –ø–æ y: {diff((2*x**2*y**3) + 1/x + (y**2*x) + 7, y)}') #–ß–∞—Å—Ç–Ω–∞—è –ø—Ä–æ–∏–∑–≤–æ–¥–Ω–∞—è –ø–æ y
```

#### –ü—Ä–æ–≤–µ—Ä–∏–º —Ä–∞—Å—á–µ—Ç —á–∞—Å—Ç–Ω—ã—Ö –ø—Ä–æ–∏–∑–≤–æ–¥–Ω—ã—Ö —Ä—É–∫–∞–º–∏:
f'x = (2*x**2*y**3)'x + (1/x)'x + (y**2*x)'x +7'x = 4*x*y**3 - 1/x**2 + y**2 + 0 =  4*x*y**3 - 1/x**2 + y**2  - —Å–æ–≤–ø–∞–ª–æ

f'y = (2*x**2*y**3)'y + (1/x)'y + (y**2*x)'y +7'y = 6*x**2*y**2 + 0 + 2*x*y + 0 = 6*x**2*y**2 + 2*x*y - —Å–æ–≤–ø–∞–ª–æ

```python
#–≤—ã—á–∏—Å–ª–∏–º –≥—Ä–∞–¥–∏–µ–Ω—Ç –≤ —Ç–æ—á–∫–µ (1;2)
a = np.array([1., 2.])
from scipy.optimize import approx_fprime #–∏–º–ø–æ—Ä—Ç–∏—Ä—É–µ–º —Ñ-—é –¥–ª—è —Ä–∞—Å—á–µ—Ç–∞ –≥—Ä–∞–¥–∏–µ–Ω—Ç–∞
```

```python
c0, c1, c2, c3 = (2, 1, 1, 7)
eps = np.sqrt(np.finfo(float).eps)
approx_fprime(a, f1, [eps, eps], c0, c1, c2, c3)
#–ø–µ—Ä–≤–∞—è –∫–æ–æ—Ä–¥–∏–Ω–∞—Ç–∞ ‚Äì —Ç–æ—á–∫–∞ –≤ –∫–æ—Ç–æ—Ä–æ–π –≤—ã—á–∏—Å–ª—è–µ–º –≥—Ä–∞–¥–∏–µ–Ω—Ç, 
#–≤—Ç–æ—Ä–∞—è ‚Äì –Ω–∞—à–∞ —Ñ—É–Ω—Ü–∏—è,
#—Ç—Ä–µ—Ç—å—è ‚Äì —ç—Ç–æ –ø–æ–≥—Ä–µ—à–Ω–æ—Å—Ç—å –≤—ã—á–∏—Å–ª–µ–Ω–∏—è, –∏ –∫–æ—ç—Ñ—Ñ–∏—Ü–∏–µ–Ω—Ç—ã –∏–∑ —Ñ—É–Ω–∫—Ü–∏–∏.
```

```python
#–ü—Ä–æ–≤–µ—Ä–∏–º —Ä–∞—Å—á–µ—Ç –≥—Ä–∞–¥–∏–µ–Ω—Ç–∞ —Ä—É–∫–∞–º–∏:
x = 1
y = 2
fx = (4*x*y**3 + y**2 - 1/x**2) #–Ω–∞—Ö–æ–¥–∏–º –∑–Ω–∞—á–µ–Ω–∏–µ —á–∞—Å—Ç–Ω–æ–π –ø—Ä–æ–∏–∑–≤–æ–¥–Ω–æ–π –ø–æ x –≤ —Ç. (2,1)
fy = (6*x**2*y**2 + 2*x*y) #–Ω–∞—Ö–æ–¥–∏–º –∑–Ω–∞—á–µ–Ω–∏–µ —á–∞—Å—Ç–Ω–æ–π –ø—Ä–æ–∏–∑–≤–æ–¥–Ω–æ–π –ø–æ y –≤ —Ç. (2,1)

grad = np.array([fx,fy])
grad
```

–†–∞—Å—á–µ—Ç –≥—Ä–∞–¥–∏–µ–Ω—Ç–∞ —Å–æ–≤–ø–∞–ª –≤—Ä—É—á–Ω—É—é –∏ —á–µ—Ä–µ–∑ —Ñ-—é

```python
# –ü–æ—Å—á–∏—Ç–∞–π—Ç–µ —á–∞—Å—Ç–Ω—ã–µ –ø—Ä–æ–∏–∑–≤–æ–¥–Ω—ã–µ —Ñ—É–Ω–∫—Ü–∏–π:2)  ùëì(ùë•,ùë¶)=ùë•2ùë¶‚àíùë†ùëñùëõ(ùë•ùë¶)+ùëêùëúùë†(ùë•2)+6ùë¶
def f2(v,c0,c1,c2,c3):
    return (c0 * v[0]**2*v[1]) + c1 * sin(v[0]*v[1]) + c2 * cos(v[1]**2) + c3*v[1]
c0,c1,c2,c3 = (1,-1,1,6)

from math import sin, cos
from sympy import *
x, y = symbols('x y')
print(f'–ß–∞—Å—Ç–Ω–∞—è –ø—Ä–æ–∏–∑–≤–æ–¥–Ω–∞—è –ø–æ —Ö: {diff((x*2*y**3) - sin(x*y) + cos(y**2) + 6*y, x)}') #–ß–∞—Å—Ç–Ω–∞—è –ø—Ä–æ–∏–∑–≤–æ–¥–Ω–∞—è –ø–æ —Ö
print(f'–ß–∞—Å—Ç–Ω–∞—è –ø—Ä–æ–∏–∑–≤–æ–¥–Ω–∞—è –ø–æ y: {diff((x*2*y**3) - sin(x*y) + cos(y**2) + 6*y, y)}')#–ß–∞—Å—Ç–Ω–∞—è –ø—Ä–æ–∏–∑–≤–æ–¥–Ω–∞—è –ø–æ y
```

–ü—Ä–æ–≤–µ—Ä–∏–º —Ä–∞—Å—á–µ—Ç —á–∞—Å—Ç–Ω—ã—Ö –ø—Ä–æ–∏–∑–≤–æ–¥–Ω—ã—Ö —Ä—É–∫–∞–º–∏:

f'x = (x*2*y**3)'x - sin(x*y)'x + cos(y**2)'x + (6*y)'x = 2*y**3 - y*cos(x*y) + 0 + 0 - —Å–æ–≤–ø–∞–ª–æ

f'y = (x*2*y**3)'y - sin(x*y)'y + cos(y**2)'y +(6*y)'y = 6*x*y**2 - x*cos(x*y) - 2*y*sin(y**2) + 6 - —Å–æ–≤–ø–∞–ª–æ

–£—Ä–æ–≤–µ–Ω—å 1:
–ì—Ä–∞–¥–∏–µ–Ω—Ç–Ω—ã–π —Å–ø—É—Å–∫ —Å–≤–æ–∏–º–∏ —Ä—É–∫–∞–º–∏:

```python
#–∑–∞–¥–∞–µ–º —Ñ-—é
def f(x,c0,c1):
    return (c0*x[0]**2) + (c1*x[1]**2)
```

```python
# –Ω–∞—á–∞–ª—å–Ω. –∑–Ω–∞—á–µ–Ω–∏–µ
c0, c1 = (1,1)
xn = np.array([100, 200])
yn = f(xn,c0, c1)
# —à–∞–≥ –ì–°
learning_rate = 0.1
```

```python
# –∑–∞–≤–æ–¥–∏–º —Å–ª–æ–≤–∞—Ä—å, –≥–¥–µ –±—É–¥–µ–º —Ö—Ä–∞–Ω–∏—Ç—å –≤—Å–µ –Ω–∞–π–¥–µ–Ω–Ω—ã–µ –∑–Ω–∞—á–µ–Ω–∏—è —Ñ—É–Ω–∫—Ü–∏–∏
Y = {yn: xn}
```

```python
Y
```

```python
for i in range(100):
    #–Ω–∞—Ö–æ–¥–∏–º –≥—Ä–∞–¥–∏–µ–Ω—Ç: 
    c0, c1 = (1,1)
    eps = np.sqrt(np.finfo(float).eps)
    grad = approx_fprime(xn, f, [eps, eps], c0, c1)
    # –ø–æ —Ñ–æ—Ä–º—É–ª–µ –≥—Ä–∞–¥–∏–µ–Ω—Ç–Ω–æ–≥–æ —Å–ø—É—Å–∫–∞ –ø–æ–ª—É—á–∞–µ–º –≤—Å–µ –∑–Ω–∞—á–µ–Ω–∏—è x y:
    xn = xn - learning_rate*grad
    yn = f(xn,c0, c1)
    Y[yn] = xn
Y
```

```python
print("–ú–∏–Ω–∏–º—É–º –¥–æ—Å—Ç–∏–≥–∞–µ—Ç—Å—è –≤: ", min(Y.items())[1])
print("–ó–Ω–∞—á–µ–Ω–∏–µ —Ñ—É–Ω–∫—Ü–∏–∏ –≤ –º–∏–Ω–∏–º—É–º–µ: ", min(Y.items())[0])
```

–ü—Ä–æ–≤–µ—Ä–∫–∞ —Å –ø–æ–º–æ—â—å—é –≤—Å—Ç—Ä–æ–µ–Ω–Ω–æ–π —Ñ—É–Ω–∫—Ü–∏–∏ numpy:

```python
def f1(x):
    return (x[0]**2) + (x[1]**2)
x0 = np.array([100, 200])

from scipy.optimize import minimize
minimize(f1, x0, method='nelder-mead', options={'xtol': 1e-8, 'disp': True})
```
