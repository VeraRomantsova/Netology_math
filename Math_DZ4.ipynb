$f(x) = sin^5(2x+1)$

На отрезке [-5; 5].

```python
#Уровень 0:
from scipy.misc import derivative
from math import sin,cos
def f(x): #значение первой производной
    return x**5 + 4*sin(2*x) + cos(3*x + 3)
derivative(f, 1.0, dx=1e-6)
```

```python
# значение первой производной руками f'(x) = (x**5 + 4*sin(2*x) + cos(3*x + 3))' = 5*x**4 - 3*sin(3*x + 3) + 8*cos(2*x)
x = 1
5*x**4 - 3*sin(3*x + 3) + 8*cos(2*x) #подставим х=1 в ф-ю производной
```

```python
def f(x): #значение второй производной
    return x**5 + 4*sin(2*x) + cos(3*x + 3)
derivative(f, 1.0, dx=1e-6, n=2)
```

```python
# значение второй производной руками f''(x) = (5*x**4 - 3*sin(3*x + 3) + 8*cos(2*x))' = 20*x**3 - 16*sin(2*x) - 9*cos(3*x + 3)
x = 1
20*x**3 - 16*sin(2*x) - 9*cos(3*x + 3) #подставим х=1 в ф-ю 2-ой производной
```

```python
#Уровень 1:
import numpy as np
from math import sin,cos
import matplotlib.pyplot as plt

def f(x):
    return sin(5*(2*x + 1))
x = np.linspace(-5, 5)
```

```python
f1x = [derivative(f, xx, dx=1e-6, n=1) for xx in x]
f2x = [derivative(f, xx, dx=1e-6, n=2) for xx in x]
```

```python
plt.plot(x, f1x, label='first derivative')
plt.plot(x, f2x, label='second derivative')
plt.legend()
plt.show()
```
